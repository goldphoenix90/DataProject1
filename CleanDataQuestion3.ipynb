```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
# Name of the CSV file
file_education = 'ACS_17_5YR_B23006_with_ann.csv'
education = pd.read_csv(file_education)

# Preview of the DataFrame
education.head()
```

```python
#select the required columns
education.iloc[:, [2, 5, 19, 33, 47]]
```

```python
#define new variable
education_clean = education.iloc[:, [2, 5, 19, 33, 47]]
```

```python
#removing header and making the first row the new header
header = education_clean.iloc[0]
education_clean = education_clean[1:]
education_clean.rename(columns = header)
```

```python
#define new variable
education_data = education_clean.rename(columns = header)
```

```python
list(education_data.columns.values)
```

```python
education_data.columns = ['Geography', 'Less than High School', 'High School Graduate', 'Associate Degree', 'Bachelor Degree or Higher']

education_data
```

```python
education_data[['Metro_Area','State']] = education_data['Geography'].str.split(',',expand=True)

education_data.head(5)
```

```python
education_data["Less than High School"] = education_data["Less than High School"].astype(float)
education_data["High School Graduate"] = education_data["High School Graduate"].astype(float)
education_data["Associate Degree"] = education_data["Associate Degree"].astype(float)
education_data["Bachelor Degree or Higher"] = education_data["Bachelor Degree or Higher"].astype(float)
```

```python
education_reduced= education_data.drop(['Geography', 'State'], axis=1)
education_reduced.set_index("Metro_Area", inplace = True)
education_reduced.head()
```

```python
highest_level_education = education_reduced.sort_values("Bachelor Degree or Higher", ascending = False)
highest_level_education.head(5)
```

```python
# New Dataset for Median Listing Price
```

```python
# Name of the CSV file
listingprice = 'Metro_MedianListingPrice_AllHomes.csv'
median_price = pd.read_csv(listingprice, encoding="latin")

# Preview of the DataFrame
median_price.head()
```

```python
def convert_to_yearly (df):
   converted = (df.groupby(pd.PeriodIndex(df.columns, freq='Y'), axis=1)
                 .mean()
                 .rename(columns=lambda c: str(c).lower()))
   return converted
```

```python
median_price_reduced= median_price.drop(['SizeRank'], axis=1)
median_price_reduced.set_index("RegionName", inplace = True)
median_price_reduced.head()
```

```python
median_price_yearly = convert_to_yearly(median_price_reduced)
median_price_yearly.head()
```

```python
final_median_price = median_price_yearly.drop(['United States'])
final_median_price.head()
```

```python
final_median_price["2010"] = final_median_price["2010"].astype(float)
final_median_price["2011"] = final_median_price["2011"].astype(float)
final_median_price["2012"] = final_median_price["2012"].astype(float)
final_median_price["2013"] = final_median_price["2013"].astype(float)
final_median_price["2014"] = final_median_price["2014"].astype(float)
final_median_price["2015"] = final_median_price["2015"].astype(float)
final_median_price["2016"] = final_median_price["2016"].astype(float)
final_median_price["2017"] = final_median_price["2017"].astype(float)
final_median_price["2018"] = final_median_price["2018"].astype(float)
final_median_price["2019"] = final_median_price["2019"].astype(float)
```

```python
final_median_price.dropna(inplace = True) 
```

```python
USA_median_price = median_price_yearly.iloc[0]
USA_median_price.plot(kind='line', linewidth=7.0, figsize=(20, 10))
plt.title("Historical Median House Price in the United States")
plt.xlabel("Year")
plt.ylabel("$ Value")
plt.savefig('median_house_price_usa.png')
```

```python
highest_median_house_price = final_median_price.sort_values("2019", ascending = False)
highest_median_house_price.head(10)
```

```python
top_5_metro_median_price = highest_median_house_price.iloc[0:5]
top_5_metro_median_price.plot(kind='bar',figsize=(15, 5))
plt.title("Top-5 Cities w/Highest Median Price in the United States")
plt.xlabel("$ Value")
plt.ylabel("Metro Area")
plt.savefig('top_5_cities_price.png')
```

```python
price_2018 = highest_median_house_price["2018"].iloc[0:5]
price_2018.plot(kind = "bar", label = "House Price",figsize=(15, 5))
plt.title("Top-5 Cities w/Highest Median Price in the United States in 2018")
plt.xlabel("Metro Area")
plt.ylabel("$ Value")
plt.savefig('top_5_cities_price_2018.png')
```
